Q3. Create 2 new columns based on Player name. First column will have first name and second column will have last name. Eg: for the player Shikhar   Dhawan, Shikhar will be the first name and Dhawan will be the last name.

```python
data['First Name'] = data['Player'].str.split().str[0]  
data['Last Name'] = data['Player'].str.split().str[1:].apply(lambda x: ' '.join(x) if x else None)  

print(data[['First Name', 'Last Name']].head())
```

 Q4. Create a new column (Cleaned_Highest_score) based on Highest score variable. Remove the Asterik(*) mark and convert the data type into INT.

```python
data['Cleaned_Highest_score'] = data['Highest Score'].str.replace('*', '', regex=False).astype(int)

print(data[['Highest Score', 'Cleaned_Highest_score']].head())
```

Q6. Print all the player names whose strike rate is less than the average strike rate of all players in entire season. Print the player name, his strike rate and average strike rate.

Q7. Please check the correlation between the features and create a heat map.

Q10. Players who scored atleast one century in this season. Create visualization.

```python
century_players = data[data['Centuries'] > 0]

century_players = century_players.sort_values(by='Centuries', ascending=False)

century_players['Color'] = century_players['Centuries'].apply(lambda x: 'More than 1 Century' if x > 1 else '1 Century')

plt.figure(figsize=(8, 4))
sns.barplot(x='Centuries', y='Player', data=century_players, hue='Color', dodge=False, palette=['yellow', 'darkviolet'])

plt.title('Players Who Scored At Least One Century', fontsize=14)
plt.xlabel('Number of Centuries', fontsize=12)

plt.tight_layout()
plt.show()
```

Q16. Print the players who played the match but didn't get the batting.

```python
players_no_batting = []

for index, row in data.iterrows():
    if row['Matches'] > 0 and row['Inns'] == 0:
        players_no_batting.append(row)

# Check if we found any such players
if players_no_batting:
    print("Players who played matches but didn't get to bat:")
    for player in players_no_batting:
        print(f"{player['Player']} -> Matches: {player['Matches']}, Not Out: {player['Not Out']}, "
              f"Inns: {player['Inns']}, Runs: {player['Runs']}, Strike Rate: {player['Strike rate']}, "
              f"Avg: {player['Avg']}, High Score: {player['Cleaned_Highest_score']}, Balls Faced: {player['Balls faced']}")
else:
    print("All players got the chance for batting")
```

Q17. Create a new column to show the percentage of total runs scored in 4s and 6s. Then print the top 5 players with maximum percentage.

```python
data['Runs_in_Fours'] = data['Most_Fours'] * 4
data['Runs_in_Sixes'] = data['Most_Sixes'] * 6
data['Total_Boundary_Runs'] = data['Runs_in_Fours'] + data['Runs_in_Sixes']

data['Boundary_Runs_Percentage'] = (data['Total_Boundary_Runs'] / data['Runs']) * 100

top_5_boundary_players = data.sort_values(by='Boundary_Runs_Percentage', ascending=False).head(5)

top_5_boundary_players[['Player', 'Runs', 'Total_Boundary_Runs', 'Boundary_Runs_Percentage']]
```

Q18. Print the players with top 5 Not out percentages (Not Out percentage can be calculated as number of Not outs divided by Innings).

```python
data['Not_Out_Percentage'] = (data['Not Out'] / data['Inns']) * 100

data['Not_Out_Percentage'].fillna(0)

top_5_not_out_players = data.sort_values(by='Not_Out_Percentage', ascending=False).head(5)

top_5_not_out_players[['Player', 'Not Out', 'Inns', 'Not_Out_Percentage']]
```
